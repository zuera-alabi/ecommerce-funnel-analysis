## Data Cleaning
- Fix event_time to datetime
- Handle missing values
- Check and drop duplicates
- Save cleaned dataset

#### Convert event_time to datetime

```python
df['event_time'] = pd.to_datetime(df['event_time'])
print(f"event_time dtype: {df['event_time'].dtype}")
```

#### Handle missing values 

```python
df = df.dropna(subset=['user_session'])
df = df.copy()
df['category_code'] = df['category_code'].fillna('Uncategorised')
df['brand'] = df['brand'].fillna('Uncategorised')

print(df.isnull().sum())
```

#### Drop duplicates 

```python
df.drop_duplicates()
```

##### Extracting year, month and day for time trends analysis 

```python
df['date'] = df['event_time'].dt.date
df['hour'] = df['event_time'].dt.hour
df['month_name'] = df['event_time'].dt.strftime('%B')
df['day_of_week'] = df['event_time'].dt.strftime('%A')

print(df[['event_time', 'date', 'hour', 'month_name', 'day_of_week']].head())
```
